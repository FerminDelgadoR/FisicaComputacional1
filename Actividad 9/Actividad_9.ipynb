***Ejercicio 3***



$$ \frac{d^2x}{dt^2}+\omega_0^2x = 0;\; \omega_0>0 $$

utilizando el cambio de variable $y=\dfrac{dx}{dt}$, se obtiene que

$$ \begin{align*} \frac{dx}{dt}&=y\\ \frac{dy}{dt}&=-\omega_0^2x \end{align*} $$

y su forma matricial seria

$$ \vec{x}'=\frac{d}{dt} \begin{pmatrix} x \\ y \end{pmatrix}=\begin{pmatrix} 0 & 1 \\ -\omega_0^2 & 0 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix}=A\vec{x}$$
por ende los eigenvalores son 

$$ \lambda=\pm \omega_0i $$

como $a=0$ y $tr(A)=0$ . se concluye que el punto critico es tipo centro

```python

W=10
array=np.array([[0,1],[W,0]])
b=np.zeros((2,1))
pcr=la.solve(array,b)
print('x =',pcr[0],'\n','y =',pcr[1])
```

entonces el punto  $(x,y)=(0,0)$ es un punto tipo centro

se mostraran ahora algunas soluciones

\frac{d y}{dt } & = &  2 z  \\

\end{eqnarray*}




```python

array = np.array([[-2,0,0],[0,0,2],[0,-2,0]])
eigenval, eigenvec = la.eig(array)
print(eigenval)
```

```python
# Resolvemos el sistema homogéneo para encontrar puntos críticos
b=np.zeros((3,1))
pcr=la.solve(array,b)
pcr
```

por lo tanto el punto (0,0,0) es un punto critico, y como la primera ecuacion donde se involucra la variable x que es independiente de las demas la solucion general se puede ver como

$$ \vec{x}=\begin{pmatrix} c_1e^{-2t} \\ c_2\cos{2t}+c_3\sin{2t} \\ -c_2\sin{2t}+c_3\cos{2t} \end{pmatrix} $$

graficando algunas soluciones

```python
# Definimos la función con la solución del sistema, como se ve arriba
def func(t,c1,c2,c3):
  EC=[c1*np.exp(-2*t), c2*np.cos(2*t)+c3*np.sin(2*t), -c2*np.sin(2*t)+c3*np.cos(2*t)]
  return EC


t=np.linspace(-5,5,100)
c1=list(np.linspace(-5,10,4))
c2=list(np.linspace(7,-2,4))
c3=list(np.linspace(-1,1,4))


plt.figure()
ax=plt.axes(projection='3d')
for c,d,e in zip(c1,c2,c3):
  sol=func(t,c,d,e)
  s=np.array(sol)
  ax.plot3D(s[0,:],s[1,:],s[2,:])
plt.xlabel('x(t)')
plt.ylabel('y(t)')
ax.set_zlabel('z(t)')

plt.show()
```
